**Applying the different formatting lines**

```python
# 1.solid line --> -
#2.dotted line --> :
#3.dashed lines --> --
#4.dash dot lines --> .
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'o:y')
plt.show()
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'o-y')
plt.show()
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'o--y')
plt.show()
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'o-.y')
plt.show()
```

```python
#increase the size of the maker from the above plot --> use 'ms='
plt.plot(a,b,'p:r',ms=10)
plt.show()
```

```python
#give the different marker edge color for above plot --> 'mec='
plt.plot(a,b,'p-.y',ms=10,mec='b')
plt.show()
```

```python
#fill the marker with the different color --> mfc=
plt.plot(a,b,'p-.y',ms=10,mec='b',mfc='#0ee68c')
plt.show()
```

```python
#plot a graph between exam entry number and the marks
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,6,6,5,8,9,10]
plt.plot(x,y,'p')
plt.xlabel('Exam Entry Number')
plt.ylabel('Marks in exam')
plt.title('Marks vs Exam Entry Number')
#use the grids()--> locate exact loction of data point
#plt.grid()
#specifiy grids according the x-axis and y-axis
#plt.grid(axis='x')
#change the linestyle of the grid
plt.grid(axis='x',linestyle='--',linewidth=1,color='#e6560e')
plt.show()
```

**scatter**

```python
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,6,6,5,8,9,10]
#draw a scatter for the above data
plt.scatter(x,y,color='r')
plt.show()
```

```python
#draw a scatter plot with different size and color
plt.scatter(x,y,color='r',s=50)
plt.show()
```

```python
x_female = [4.6,5.2,5.4,4.9,5.1,5.8,4.8]
y_female = [63,40,35,58,54,56,58]
plt.scatter(x_female,y_female,color='#e60edf',s=50)
plt.show()
```

```python
#change the size of scatters according to the wight of female
x_female = [4.6,5.2,5.4,4.9,5.1,5.8,4.8]
y_female = [63,40,35,58,54,56,58]
plt.scatter(x_female,y_female,color='#e60edf',s=y_female)
plt.show()
```
