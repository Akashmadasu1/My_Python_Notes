**Applying the different formatting lines**

```python
# 1.solid line --> -
#2.dotted line --> :
#3.dashed lines --> --
#4.dash dot lines --> .
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'o:y')
plt.show()
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'o-y')
plt.show()
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'o--y')
plt.show()
```

```python
a = np.array([1,2,3,4,5,6])
b = np.array([3,8,1,5,10,7])
plt.plot(a,b,'o-.y')
plt.show()
```

```python
#increase the size of the maker from the above plot --> use 'ms='
plt.plot(a,b,'p:r',ms=10)
plt.show()
```

```python
#give the different marker edge color for above plot --> 'mec='
plt.plot(a,b,'p-.y',ms=10,mec='b')
plt.show()
```

```python
#fill the marker with the different color --> mfc=
plt.plot(a,b,'p-.y',ms=10,mec='b',mfc='#0ee68c')
plt.show()
```

```python
#plot a graph between exam entry number and the marks
x = [1,2,3,4,5,6,7,8,9]
y = [2,1,4,6,6,5,8,9,10]
plt.plot(x,y,'p')
plt.xlabel('Exam Entry Number')
plt.ylabel('Marks in exam')
plt.title('Marks vs Exam Entry Number')
#use the grids()--> locate exact loction of data point
#plt.grid()
#specifiy grids according the x-axis and y-axis
#plt.grid(axis='x')
#change the linestyle of the grid
plt.grid(axis='x',linestyle='--',linewidth=1,color='#e6560e')
plt.show()
```

#Seaborn

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
#use the default datasets from the seaborn
data = sns.load_dataset('tips')
data.head()
```

```python
#draw a line plot between the tip and total_bill
plt.plot(data['tip'],data['total_bill'],'o:r')
plt.show()
```

```python
#draw a scatter plot between 'sex' and total_bill
plt.scatter(data['sex'],data['total_bill'])
plt.show()
```

```python
plt.scatter(data['tip'],data['total_bill'])
plt.show()
```

**scatter plot by using seaborn**

```python
sns.scatterplot(data = data,x='total_bill',y='tip')
plt.show()
```

```python
data.columns
```

give the colours based on category ----> hue

```python
#give the colours based on category ----> hue
#draw a sctterplot b/w total_bill vs tip based on category smoker
sns.scatterplot(data = data,x='total_bill',y='tip',hue='smoker')
plt.show()
```

```python
#draw a sctterplot b/w total_bill vs tip based on category 'day'
sns.scatterplot(data = data,x='total_bill',y='tip',hue='day')
plt.show()
```

```python
#draw a sctterplot b/w total_bill vs tip based on category 'day' and size
sns.scatterplot(data = data,x='total_bill',y='tip',hue='day',size='size')
plt.show()
```

```python
#draw a sctterplot b/w total_bill vs tip based on category 'day' and size of the people and on what time bill and tip has paid
sns.scatterplot(data = data,x='total_bill',y='tip',hue='day',size='size',style='time')
plt.show()
```

```python
#how much total_bill has been given by different size group of people on different day?
sns.scatterplot(data = data,x='tip',y='total_bill',hue='sex',size='size',style='day')
plt.show()
```
